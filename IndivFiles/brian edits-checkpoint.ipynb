```python
#import dependenceis
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import datetime
```

```python
#view data sources

clin_data_path = '../OriginalData/WHO_NREVSS_Clinical_Labs.csv'

#read data sources
clin_data = pd.read_csv(clin_data_path, skiprows = 1)

```

```python
#view clin data
clin_df = pd.DataFrame(clin_data)


clin_df.rename(columns={'TOTAL SPECIMENS': 'TOTAL'}, inplace=True)

clin_df.head(10)
```

```python
#view clin data
clin_df = pd.DataFrame(clin_data)

#remove region type

clin_drop = clin_df.drop(columns = 'REGION TYPE')

#remove rows containing X
clin_drop2 = clin_drop[clin_drop['TOTAL'] != 'X']
clin_drop2 = clin_drop[clin_drop['TOTAL A'] != 'X']
clin_drop2 = clin_drop[clin_drop['TOTAL B'] != 'X']
clin_drop2 = clin_drop[clin_drop['PERCENT POSITIVE'] != 'X']
clin_drop2 = clin_drop[clin_drop['PERCENT A'] != 'X']
clin_drop2 = clin_drop[clin_drop['PERCENT B'] != 'X']

#convert datatypes of columns to float
clin_drop2 = clin_drop2[['REGION', 'WEEK', 'YEAR','TOTAL', 'TOTAL A', 'TOTAL B', 'PERCENT POSITIVE', 'PERCENT A', 'PERCENT B']]

df_subset = clin_drop2

for column in df_subset.columns[3:]:
    df_subset[column] = df_subset[column].astype('float')


#clean clin
nxt_clin = df_subset

nxt_clin.head (10)
```

```python
clin_max = nxt_clin['TOTAL'].sum()
clin_max = nxt_clin['TOTAL A'].sum()
clin_max = nxt_clin['TOTAL B'].sum()
```

```python
weeks = nxt_clin['WEEK']

season = []

for week in weeks:
    if week >= 51:
        season.append('WINTER')
    elif week >= 38:
        season.append('FALL')
    elif week >= 25:
        season.append('SUMMER')
    elif week >= 12:
        season.append('SPRING')
    elif week >=1:
        season.append('WINTER')
        
nxt_clin['SEASON'] = season
    
nxt_clin.head()
```

```python
nxt_clin.info()
```

```python
season_df = nxt_clin.groupby('SEASON').sum()
#season_df = nxt_clin.groupby('SEASON').sum()['TOTAL A']
#season_df = nxt_clin.groupby('SEASON').sum()['TOTAL B']
season_df.head()
```

```python
nxt_clin.to_csv('clinical.csv',index=False)
```

```python
#Clin bar graph: x-axis = season (weeks), y-axis = num total positives 

plt.bar(season_df.index.values, season_df['TOTAL'], color='blue')
plt.xticks(rotation=90)
plt.xlabel('Season')
plt.ylabel('Total Influenza positives (millions)')
plt.title('Total Influenza infections per season (U.S.)')
plt.show()
```

```python
#Clin bar graph: comparing TOTAL A and TOTAL B Influenza

# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

x = season_df.index.values
y = season_df['TOTAL A']
z = season_df['TOTAL B']
    
# Make the plot
plt.bar(x, y, color ='r', width = barWidth,
        edgecolor ='grey', label ='Total A')
plt.bar(x, z, color ='g', width = barWidth,
        edgecolor ='grey', label ='Total B')

 
# Adding Xticks
plt.xlabel('SEASONS', fontweight ='bold', fontsize = 15)
plt.ylabel('POSITIVE INFLUENZA STRAINS of A & B', fontweight ='bold', fontsize = 15)
plt.xticks(rotation=90)
 
plt.legend()
plt.show()

```
