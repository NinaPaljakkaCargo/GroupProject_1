```python
#view clin data
clin_df = pd.DataFrame(clin_data)

#remove region type

clin_drop = clin_df.drop(columns = 'REGION TYPE')

#remove rows containing X
clin_drop2 = clin_drop[clin_drop['TOTAL SPECIMENS'] != 'X']
clin_drop2 = clin_drop[clin_drop['TOTAL A'] != 'X']
clin_drop2 = clin_drop[clin_drop['TOTAL B'] != 'X']
clin_drop2 = clin_drop[clin_drop['PERCENT POSITIVE'] != 'X']
clin_drop2 = clin_drop[clin_drop['PERCENT A'] != 'X']
clin_drop2 = clin_drop[clin_drop['PERCENT B'] != 'X']

#convert datatypes of columns to float
clin_drop2 = clin_drop2[['REGION', 'WEEK', 'YEAR','TOTAL SPECIMENS', 'TOTAL A', 'TOTAL B', 'PERCENT POSITIVE', 'PERCENT A', 'PERCENT B']]

#clean clin
nxt_clin = clin_drop2

nxt_clin.head (10)
```

```python
weeks = nxt_clin['WEEK']
clin_max = nxt_clin['TOTAL SPECIMENS'].sum()

clin_spring = 0
clin_summer = 0
clin_fall = 0
clin_winter = 0

for row in weeks:
    if row >= 12 & row <= 24:
        clin_total = nxt_clin['TOTAL SPECIMENS']
        clin_spring = clin_spring + clin_total[12]
    if row >= 25 & row <= 37:
        clin_total = nxt_clin['TOTAL SPECIMENS']
        clin_summer = clin_summer + clin_total[25]  
    if row >= 38 & row <= 50:
        clin_total3 = nxt_clin['TOTAL SPECIMENS']
        clin_fall = clin_fall + clin_total3[38]
    #if row == 51 | row == 52:
        #clin_total = nxt_clin['TOTAL SPECIMENS']
        #clin_winter = clin_winter + clin_total
    if row >= 1 & row <= 12:
        clin_total = nxt_clin['TOTAL SPECIMENS']
        clin_winter = clin_winter + clin_total[1]


print(clin_spring)
print(clin_summer)
print(clin_fall)
print(clin_winter)
print(clin_max)
```
