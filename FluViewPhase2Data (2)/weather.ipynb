```python
# Dependencies
import requests
import json
```

```python
# URL for GET requests to retrieve vehicle data
# https://www.ncdc.noaa.gov//homr/services/station/search?state=CA&platform=ASOS
state = 'CA'
url = f"https://www.ncdc.noaa.gov//homr/services/station/search?state={state}&platform=ASOS"
```

```python
response_json = requests.get(url)
#json_dumps = json.dumps(response_json, indent=2, sort_keys=True)
print(response_json)
```

```python
print(response_json.content)
```

```python
json_loads = json.loads(response_json.text)
print(json_loads.keys())
```

```python
print(json_loads['stationCollection']['stations'])
```

```python
for stations in json_loads['stationCollection']['stations']:
    print(stations['identifiers'][0]['id'])
    try:
        print(stations['location']['nwsInfo']['climateDivisions'][0]['displayName'])
    except:
        print(f'Station name not found')

```
